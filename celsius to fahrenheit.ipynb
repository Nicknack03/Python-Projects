```python
import tensorflow.compat.v1 as tf #converting tensorflow version 2 into version 1 to show logging attribute
```

```python
tf.logging.set_verbosity(tf.logging.ERROR) # to display message of error
```

```python
import numpy as np
```

```python
celsius_q    = np.array([-40, -10, 0, 8, 15, 22,  38], dtype=float)
fahrenheit_a = np.array([-40, 14, 32, 46, 59, 72, 100], dtype=float)

for i,c in enumerate(celsius_q): #enumerate keeps the count of iterations
  print(str(c)+" degrees in Celsius = "+str(fahrenheit_a[i])+" degrees Fahrenheit")
```

```python
lO = tf.keras.layers.Dense(units=1, input_shape=[1]) #input_shape means that the input to this layer is a single value
                                                     #unit means number is neurons in the layer
```

```python
model = tf.keras.Sequential([lO])    #The sequential API allows you to create models layer-by-layer for most problems. It is limited in that it does not allow you to create models that share layers or have multiple inputs or outputs.
                                     #keras is used to make the implementation of neural networks easy
```

```python
model.compile(loss='mean_squared_error',                    
              optimizer=tf.keras.optimizers.Adam(0.1)) #0.1 is the learning rate 
#tensorflow uses these two function to find the best model during training.
```

```python
history = model.fit(celsius_q, fahrenheit_a, epochs=500, verbose=False) #epochs means exapmle or iteration
print("finished training the model")
```

```python
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
```

```python
print(model.predict([0.0]))
```
